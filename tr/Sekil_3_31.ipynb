```python
# Tanimlar
def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
import matplotlib.pyplot as plt
import numpy as np
import pywt   
get_ipython().magic('matplotlib inline')

#Meksika Sapkasi
mexh=pywt.ContinuousWavelet('mexh')
psi,x=mexh.wavefun(level=10)

setup_graph(x_label='Ornek', y_label='Genlik', title=u'Meksika Şapkası')
plt.plot(x, psi, 'k');
plt.ylim([-2,2]);

#Shannon
shan=pywt.ContinuousWavelet('shan')
psi,x=shan.wavefun(level=10);
setup_graph(x_label='Ornek', y_label='Genlik', title=u'Shannon')
plt.plot(x, psi, 'k');
plt.ylim([-2,2]);

#HAAR
haar=pywt.Wavelet('haar')
phi, psi, x = haar.wavefun(level=10);
setup_graph(x_label='Ornek', y_label='Genlik', title=u'Haar')
plt.plot(x, psi, 'k');plt.plot(x, phi, 'r');
plt.ylim([-2,2]);

#DB2
db2=pywt.Wavelet('db2')
phi, psi, x = db2.wavefun(level=10);
setup_graph(x_label='Ornek', y_label='Genlik', title=u'Daubechies-2')
plt.plot(x, psi, 'k');plt.plot(x, phi, 'r');
plt.ylim([-2,2]);
```
