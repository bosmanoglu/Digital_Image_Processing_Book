### Sekil 3.22


```python
# Tanimlar
def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc    

## DTMF Tonlari: 697,770,852,941 Hz ve 1209,1336,1477,1633 Hz
t=np.linspace(0,1,8000) # ornekleme araligi ornekleyecegimiz frekansin en az iki kati olmali.
ton1=np.sin(2*np.pi*697*t)

setup_graph(x_label='Zaman (sn)', y_label='Genlik', title='697 Hz Isaret (Zaman)')
plt.plot(t[0:100],ton1[0:100], 'k');plt.gray();
plt.xlim([0,0.01]);
```

### Frekans 

```python
Fourier_ton1=np.fft.fft(ton1)*2/8000;
freq = np.fft.fftfreq(t.shape[-1], d=np.diff(t).mean())

setup_graph(x_label='Frekans (Hz)', y_label='Genlik (dB)', title='697 Hz Isaret (Frekans)')
plt.plot(freq[0:4000],20*np.log10(abs(Fourier_ton1[0:4000])), 'k');
plt.xlim([0,1000]);plt.ylim([-100,0])
plt.show()
```
