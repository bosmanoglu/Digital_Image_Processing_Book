```python
# coding: utf-8
# Graphing helper function
def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    
```

```python
get_ipython().magic('matplotlib inline')
import matplotlib.pyplot as plt
import numpy as np
import scipy.misc
```

```python
aircam=scipy.misc.imread('../ext/airborne_image.tif', flatten=True)
aircam=scipy.misc.imresize(aircam, (100,100))
freq=np.linspace(-50,50,100)
#racoon=scipy.misc.face(gray=True)
#racoon=scipy.misc.imresize(racoon, A.shape)
setup_graph(x_label='X', y_label='Y', title=u'Hava Grnts - Mekansal')
plt.imshow(aircam);plt.gray();plt.colorbar();

Faircam=np.fft.fftshift(np.fft.fft2(aircam));
setup_graph(x_label='X', y_label='Y', title=u'Hava Grnts  - Frekans Genlik')
plt.imshow(10*np.log10(abs(Faircam)), extent=[freq[0], freq[-1], freq[-1], freq[0]]);plt.gray();plt.colorbar();

setup_graph(x_label='X', y_label='Y', title=u'Hava Grnts  - Frekans Faz')
plt.imshow((np.angle(Faircam)), extent=[freq[0], freq[-1], freq[-1], freq[0]]);plt.gray();plt.colorbar();
```
